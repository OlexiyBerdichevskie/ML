```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mtx = df.drop('DEATH_EVENT', axis=1).corr(numeric_only=True).abs()

# Побудова теплової карти кореляції
fig, ax = plt.subplots(figsize=(8, 8))

sns.heatmap(
    mtx,
    cmap='crest',
    annot=True,
    fmt=".2f",
    linewidths=0.5,
    mask=np.triu(np.ones_like(mtx, dtype=bool)),  # маскуємо верхній трикутник
    square=True,
    cbar=False,
    ax=ax
)

plt.title("Матриця попарної кореляції ознак", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
corr_with_EVENT = df.corr()['DEATH_EVENT'].abs().sort_values(ascending=False)
corr_with_EVENT
```

```python
df['DEATH_EVENT'].value_counts()
```

```python
from sklearn.feature_selection import mutual_info_classif
```

```python
from sklearn.feature_selection import mutual_info_classif

x = df.drop(columns=['DEATH_EVENT'])
y = df['DEATH_EVENT']

discrete_features = x.columns.isin([
    'anaemia', 'diabetes', 'high_blood_pressure',
    'sex', 'smoking'
])

mi_scores = mutual_info_classif(
    x, y,
    discrete_features=discrete_features,
    random_state=42
)

mi_scores = pd.Series(
    mi_scores,
    name='MI Scores',
    index=x.columns
).sort_values()

print(mi_scores.sample(5))
```

```python
nonzero_mi_scores = mi_scores[mi_scores > 0].sort_values(ascending=False)
top_mi_scores = nonzero_mi_scores.head(7)
plt.figure(figsize=(8,6))
plt.barh(top_mi_scores.index[::-1], top_mi_scores[::-1])
plt.title('Top Mantual Onformation Scores')
plt.xlabel('Score')
plt.tight_layout()
plt.show()
```
